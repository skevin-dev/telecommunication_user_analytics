```python
col_used = ['Dur. (ms).1','Social Media DL (Bytes)','Social Media UL (Bytes)','Youtube DL (Bytes)','Youtube UL (Bytes)','Netflix DL (Bytes)','Netflix UL (Bytes)','Google DL (Bytes)','Google UL (Bytes)','Email DL (Bytes)','Email UL (Bytes)','Gaming DL (Bytes)','Gaming UL (Bytes)','Total DL (Bytes)','Total UL (Bytes)']
```

```python
df_col_used = df[col_used]
```

```python
df_col_used['Social Media Total Data(Bytes)'] = df_col_used['Social Media DL (Bytes)'] + df_col_used['Social Media UL (Bytes)']
df_col_used['Youtube Total Data(Bytes)'] = df_col_used['Youtube DL (Bytes)'] + df_col_used['Youtube UL (Bytes)']
df_col_used['Netflix Total Data(Bytes)'] = df_col_used['Netflix DL (Bytes)'] + df_col_used['Netflix UL (Bytes)']
df_col_used['Google Total Data(Bytes)'] = df_col_used['Google DL (Bytes)'] + df_col_used['Google UL (Bytes)']
df_col_used['Email Total Data(Bytes)'] = df_col_used['Email DL (Bytes)'] + df_col_used['Email UL (Bytes)']
df_col_used['Gaming Total Data(Bytes)'] = df_col_used['Gaming DL (Bytes)'] + df_col_used['Gaming UL (Bytes)']
```

```python
#non-graphical analysis
df_used = df_col_used[['Dur. (ms).1','Social Media Total Data(Bytes)','Youtube Total Data(Bytes)','Netflix Total Data(Bytes)','Google Total Data(Bytes)',
                     'Email Total Data(Bytes)','Gaming Total Data(Bytes)','Total UL (Bytes)','Total DL (Bytes)']]
```

###### Graphical analysis(Univariate)

```python
for i in df_used.iloc[:,:-1]:
    sns.displot(df_used, x=i).set(title=i)
```

###### Graphical analysis(bivariate) 

```python
df_used['total'] = df_used['Total UL (Bytes)'] + df_used['Total DL (Bytes)'] 
```

```python
e = df_used.columns.difference(['Dur. (ms).1','Total UL (Bytes)','Total DL (Bytes)']).tolist()
df_used_ = df_used[e]
```

```python
df_used_
```

```python
for i in df_used_.iloc[:,:-1].columns.tolist():
    sns.scatterplot(x=df_used_[i],y=df_used_['total'],label=i)
    plt.show()
```
